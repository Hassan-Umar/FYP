```python
import csv

reviews=[]
for row in csv.reader(open('C:\\Users\\Hassan\\Downloads\\datasets.csv', encoding = "utf-8")):
    reviews.append(row)
reviews = reviews[1:]  #This command removes the first row of the file as it only contains label(Heading)
print(reviews)
```

```python
import spacy
from spacy.lang.ur import Urdu
parser = Urdu()
from spacy.lang.ur.stop_words import STOP_WORDS
from nltk.stem.wordnet import WordNetLemmatizer


final=[]     #In this list all of our pre processed data will be stored
for line in reviews:  #Break sentence into list words
    for word in line: #Break list of words into sentence
        token_list = []
        line = parser(word)  #parsing the sentence using spacy
        for token in line:   #breaking sentence into words
            if token.is_stop == False and len(token)>3:  #checking if the word is not a stopword or have less than 3 letters
                token_list.append(WordNetLemmatizer().lemmatize(token.text)) #if not then lemmatize it and add it to list of tokens
        print(token_list)
    final.append(token_list)   #append all the list into the final  
#print(final)

```

```python
from gensim import corpora
dictionary = corpora.Dictionary(final)
print(dictionary)
```

```python
corpus = [dictionary.doc2bow(text) for text in final]
print(corpus)
```
